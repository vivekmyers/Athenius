```python
import user_score

def permute(x):
    if x == 0: return [[]]
    else: return [[-1] + x for x in permute(x-1)] + [[1] + x for x in permute(x-1)]
    
permute(3)
```

```python
import random
p = permute(16)
random.shuffle(p)

for i in p:
    x, y = user_score.nearestNeighbors(i)
    a = x[0][0]
    b = x[1][0]
    c = y[0][0]
    if not all(['(R' in x for x in [a, b, c]]) and not all(['(D' in x for x in [a, b, c]]):
        print(i)
        print(a, b, c)
    
```
